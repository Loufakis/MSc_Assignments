## Cleaning

In the next section of preprocessing, it is important to apply feature cleaning in order to remove any symbols and ensure that the values are in a clean format. This is crucial because having clean and standardized data is essential for the success of any data analysis or machine learning model. Cleaning the data ensures that it is consistent and easy to work with, reducing the risk of errors and improving the accuracy of results. Additionally, removing any extraneous symbols or characters helps to reduce the dimensionality of the data, making it more manageable for analysis. By ensuring that the data is in a clean format, we can increase the reliability and interpretability of the results.

#### Feature: "Country_Code"

```python
# Defeine cleaning function
def only_alpha_2_code(string):
    return string[0:2]

# Apply
data['Country_Code'] = data['Country_Code'].apply(only_alpha_2_code)
```

#### Feature: "Topic"

```python
# Defeine cleaning function
def only_wav_code(string):
    return 'VAW_'+(string.split(':')[0][-3:])


# save info
topic_info = {}

for value in data['Topic'].unique():
    s = value.split(':')
    topic_info[s[0]] = s[1][1:]

# Apply
data['Topic'] = data['Topic'].apply(only_wav_code)
```

#### Feature: "Age"

```python
# Defeine cleaning function
def age_categ(string):
    return '15-64'

# Apply
data['Age'] = data['Age'].apply(age_categ)
```

#### Features: "Condition", "Violence_Type", "Perpetrator", "Actuality", "Life_Period", "Outcome", "Response", "Help_Reason", "Help_Provider"

```python
# Defeine cleaning function
def only_code(string):
    string = string.split(':')[0]
    if string == '_T':
        string = 'ANY'
    return string

cols_to_chng = ['Condition','Violence_Type','Perpetrator','Actuality','Life_Period',
                'Outcome','Response','Help_Reason','Help_Provider']

cols_info_dict = {}

# Apply
for col in cols_to_chng:
    # save info
    cols_info_dict[col] = {}

    for value in data[col].unique():
        s = value.split(':')
        if s[0] == '_T':
            s[0] = 'ANY'
        cols_info_dict[col][s[0]] = s[1][1:]

    data[col] = data[col].apply(only_code)
```

### Encodings

#### Greate new columns for Age (Ajustments later)

```python
col_loc = data.shape[1]

for column in ['Age_15-49', 'Age_18-64', 'Age_15-64']:
    data.insert(col_loc, f'{column}', 0, True)
    
data['Age_15-64'] = 1
        
data = data.drop('Age', axis=1)
```

#### One hot Encoding for categorical columns

```python
# columns to encode
columns = ['Country_Code','Topic', 'Condition', 'Violence_Type', 'Perpetrator', 'Actuality', 'Life_Period', 'Data_Source']
prefix  = ['Country','Topic','Cndtn','Vio_typ','Prptr','Actual','L_Per','D_S']

for idx, column in enumerate(columns):
    dummies = pd.get_dummies(data[column], prefix=prefix[idx]) #Get only the encoded columns (as a DF)
    data = pd.concat([data, dummies], axis=1) #Concat the 2 DF's
    data = data.drop(column, axis=1) #Drop the original column
```

#### Fill columns by feature "Comment"

```python
rows = data.shape[0]

def fill_cols(col_dict, prexif):
    # iterate through rows
    for row in range(rows):

        # Check if there is an avaliable comment in this row
        if str(data['Comment'].iloc[row]) != 'nan':

            # iterate through keywords
            for keyword in col_dict:

                # Check if the keyword exists in the comment
                if keyword in data['Comment'].iloc[row]:
                    change_any = True
                    
                    # iterate through columns to be adjust
                    for col in col_dict[keyword]:

                            data[f'{prexif}_{col}'].iloc[row] = 1
                            
                            if col == 'ANY':
                                change_any = False
                                
                    if change_any:
                        data[f'{prexif}_ANY'].iloc[row] = 0
                    
fill_cols(condition, 'Cndtn')
fill_cols(perpetrator, 'Prptr')
fill_cols(violence_type, 'Vio_typ')
```

#### Filling feature: "Age"

```python
rows = data.shape[0]

for row in range(rows):
    for key_word in age:
        if str(data['Comment'].iloc[row]) != 'nan':
            if key_word in data['Comment'].iloc[row]:
                data[f'Age_{age[key_word][0]}'].iloc[row] = 1
                data['Age_15-64'].iloc[row] = 0
```

#### Filling feature: "Period of Life"

```python
rows = data.shape[0]
life_per
for row in range(rows):
    
    if str(data['Comment'].iloc[row]) != 'nan':
        
        for keyword in life_per:
        
            if keyword in data['Comment'].iloc[row] and data['Age_18-64'].iloc[row] == 1:
                
                data[f'L_Per_{life_per[keyword][0]}'].iloc[row] = 1
                data['L_Per_ANY'].iloc[row] = 0
```

#### Filling feature: "Perpetrator"

```python
# iterate through rows
for row in range(rows):

    # Check if there is an avaliable comment in this row
    if str(data['Comment'].iloc[row]) != 'nan':

        # iterate through keywords
        for keyword in topic:
            go_on = True
            
            # Check if the keyword exists in the comment
            if keyword in data['Comment'].iloc[row]:
                
                if keyword == 'physical or sexual violence':
                    
                    if data['Prptr_PARTNER'].iloc[row] == 1:
                        cols_dict = topic[keyword]['partnr']
                    
                    elif data['Prptr_PARTNER'].iloc[row] != 1 and data['Prptr_ANY'].iloc[row] != 1:
                        cols_dict = topic[keyword]['n-partnr']
                        
                    else:
                        go_on = False
                
                else:
                    cols_dict = topic[keyword] 
                     
                # iterate through columns to be adjust
                if go_on:
                    for col in cols_dict:

                            data[f'Topic_{col}'].iloc[row] = 1
```

```python
def heatmaps_side_by_side(df1, df2, title, cmap):

    fig, (ax1, ax2) =plt.subplots(nrows=1,ncols=2, figsize=(18,13))
    fig.suptitle(title)
    
    sns.heatmap(df1,
                annot=True,
                norm=LogNorm(),
                cmap=cmap,
                fmt='.0f',
                ax=ax1
               )
    sns.heatmap(df2,
                annot=True,
                norm=LogNorm(),
                cmap=cmap,
                fmt='.0f',
                ax=ax2
               )
    plt.show()
```

```python
old_cols = ['Age','Country_Code','Condition','Perpetrator',
            'Violence_Type','Life_Period','Topic',
            'Actuality','Data_Source']

encod_cols = [age_cols, coutry_cols, condition_cols, perpetrator_cols, 
              violence_type_cols, life_period_cols, topic_cols, 
              actuality_cols, data_source_cols]

titles = ['Age', 'Country', 'Condition','Perpetrator',
          'Violence_Type','Life_Period','Topic',
          'Actuality','Data_Source']

for idx in range(len(encod_cols)):
    
    # Crowwtabulation Matrix
    tab1 = pd.crosstab(vaw_v1['Outcome'], vaw_v1[old_cols[idx]])
    
    # Multi-column frequency count
    tab2 = vaw_v2.groupby(['Outcome'])[encod_cols[idx]].sum()
    
    heatmaps_side_by_side(tab1, tab2, titles[idx], 'RdYlGn')
```
